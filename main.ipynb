# plotlyのテスト

plotlyライブラリの動作確認jupyter

```python
import numpy as np
import plotly.graph_objects as go

# figure widgetの作成

fig = go.FigureWidget()
fig
```

```python
# figure widgetにデータを追加する。
n: int = 100
x: np.ndarray = np.linspace(0, 20, n)
y: np.ndarray = np.random.rand(n)

fig.add_scatter(x=x, y=y)
```

```python
# figure widgetのデータを更新する

# 散布図データ参照
scatter_data = fig.data[0]
scatter_data.y = np.random.rand(n)
fig
```

```python
# 複数回更新（アニメーション）
update_num: int = 1000
scatter_data = fig.data[0]

fig

for i in range(update_num):
    scatter_data.y = np.random.rand(n)
    fig.update_layout(title=f"更新回数{i+1}/{update_num}")
```
